```python
from sklearn.model_selection import train_test_split

train_texts, test_texts, train_labels, test_labels = train_test_split(
    df["text"].tolist(),
    df["label_id"].tolist(),
    test_size=0.2,
    random_state=42
)

print("Train Samples:", len(train_texts))
print("Test Samples:", len(test_texts))
```

```python
from transformers import DistilBertTokenizer

tokenizer = DistilBertTokenizer.from_pretrained(
    "distilbert-base-uncased"
)
```

```python
train_encodings = tokenizer(
    train_texts,
    truncation=True,
    padding=True,
    max_length=128
)

test_encodings = tokenizer(
    test_texts,
    truncation=True,
    padding=True,
    max_length=128
)
```

```python
import torch

class ClauseDataset(torch.utils.data.Dataset):

    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):

        item = {
            key: torch.tensor(val[idx])
            for key, val in self.encodings.items()
        }

        item["labels"] = torch.tensor(
            self.labels[idx]
        )

        return item

    def __len__(self):
        return len(self.labels)
```

```python
train_dataset = ClauseDataset(
    train_encodings,
    train_labels
)

test_dataset = ClauseDataset(
    test_encodings,
    test_labels
)
```

```python
pdf = train_data[1]["pdf"]

full_text = ""

for page in pdf.pages:

    page_text = page.extract_text()

    if page_text:
        full_text += page_text + "\n"

print("Characters:", len(full_text))
```

```python
print(full_text[:10000])
```

```python
import re

headings = re.findall(
    r"\n\d+\.\s+[A-Za-z].+",
    full_text
)

print("Total Headings:", len(headings))

for h in headings[:20]:
    print(h)
```

```python
import re

pattern = r"\n(\d+\.\s+[^\n]+)"

matches = list(re.finditer(pattern, full_text))

print("Sections Found:", len(matches))
```

```python
clauses = []

for i in range(len(matches)):

    start = matches[i].start()

    title = matches[i].group(1)

    if i < len(matches) - 1:
        end = matches[i + 1].start()
    else:
        end = len(full_text)

    section_text = full_text[start:end]

    clauses.append({
        "label": title,
        "text": section_text
    })

print("Total Clauses:", len(clauses))
```

```python
for clause in clauses[:3]:

    print("="*80)

    print("LABEL:")
    print(clause["label"])

    print("\nTEXT:")
    print(clause["text"][:500])
```

```python
def assign_category(label):

    label = label.lower()

    if any(x in label for x in [
        "payment",
        "compensation",
        "wage",
        "fee",
        "escrow",
        "purchase price",
        "remuneration"
    ]):
        return "Payment"

    elif any(x in label for x in [
        "confidential",
        "non disclosure"
    ]):
        return "Confidentiality"

    elif any(x in label for x in [
        "leave",
        "employee",
        "employment",
        "clothing",
        "accommodation"
    ]):
        return "Employment"

    elif any(x in label for x in [
        "warranty",
        "indemnification",
        "liability",
        "damages"
    ]):
        return "Liability"

    elif any(x in label for x in [
        "premises",
        "deed",
        "property"
    ]):
        return "Property"

    elif any(x in label for x in [
        "law",
        "dispute",
        "agreement",
        "jurisdiction"
    ]):
        return "Legal"

    elif any(x in label for x in [
        "copyright",
        "trademark",
        "work products",
        "license"
    ]):
        return "Intellectual_Property"

    elif any(x in label for x in [
        "duties",
        "performance",
        "reports",
        "commencement"
    ]):
        return "Performance"

    elif any(x in label for x in [
        "termination",
        "default"
    ]):
        return "Termination"

    else:
        return "General"
```

```python
df["category"] = df["label"].apply(
    assign_category
)

print(
    df["category"].value_counts()
)
```
