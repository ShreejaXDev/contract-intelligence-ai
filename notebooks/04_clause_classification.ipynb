```python
contracts = []

for i in range(10):

    pdf = train_data[i]["pdf"]

    text = ""

    for page in pdf.pages:

        page_text = page.extract_text()

        if page_text:
            text += page_text + "\n"

    contracts.append(text)

print("Loaded:", len(contracts))
```

```python
import pandas as pd

data = {
    "text": [
        "Either party may terminate this agreement with 30 days notice.",
        "All confidential information shall remain secret.",
        "Payment shall be made within 30 days.",
        "Employee shall undergo probation for six months.",
        "Employee is entitled to 20 days annual leave.",

        "The agreement can be terminated immediately for breach.",
        "Confidential documents must not be disclosed.",
        "Invoices shall be paid monthly.",
        "Probation period shall be three months.",
        "Employees may avail paid leave."
    ],

    "label": [
        "Termination",
        "Confidentiality",
        "Payment",
        "Probation",
        "Leave Policy",

        "Termination",
        "Confidentiality",
        "Payment",
        "Probation",
        "Leave Policy"
    ]
}

df = pd.DataFrame(data)

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df["label_id"] = encoder.fit_transform(df["label"])

df
```

```python
print(dict(zip(
    encoder.classes_,
    encoder.transform(encoder.classes_)
)))
```

```python
from sklearn.model_selection import train_test_split

train_texts, test_texts, train_labels, test_labels = train_test_split(
    df["text"].tolist(),
    df["label_id"].tolist(),
    test_size=0.2,
    random_state=42
)

print("Train Samples:", len(train_texts))
print("Test Samples:", len(test_texts))
```

```python
from transformers import DistilBertTokenizer

tokenizer = DistilBertTokenizer.from_pretrained(
    "distilbert-base-uncased"
)
```

```python
train_encodings = tokenizer(
    train_texts,
    truncation=True,
    padding=True,
    max_length=128
)

test_encodings = tokenizer(
    test_texts,
    truncation=True,
    padding=True,
    max_length=128
)
```

```python
import torch

class ClauseDataset(torch.utils.data.Dataset):

    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):

        item = {
            key: torch.tensor(val[idx])
            for key, val in self.encodings.items()
        }

        item["labels"] = torch.tensor(
            self.labels[idx]
        )

        return item

    def __len__(self):
        return len(self.labels)
```

```python
train_dataset = ClauseDataset(
    train_encodings,
    train_labels
)

test_dataset = ClauseDataset(
    test_encodings,
    test_labels
)
```
